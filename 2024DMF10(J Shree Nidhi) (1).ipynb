```python
#STEP 3 - CLASSES

import pandas as pd

class ESGAnalyzer:
    def __init__(self, file_path):
        """Load the dataset."""
        self.df = pd.read_csv(file_path)
    
    def get_esg_info(self, company_name):
        """Retrieve ESG details for a given company name."""
        company_data = self.df[self.df['Name'].str.lower() == company_name.lower()]
        
        if company_data.empty:
            return "Company not found in the dataset. Please check the spelling."
        
        grade = company_data['Total_grade'].values[0]
        score = company_data['Total_score'].values[0]
        level = company_data['Total_level'].values[0]
        
        
        if score >= 1100:
            insight = "Excellent ESG performance! The company is a sustainability leader."
        elif 900 <= score < 1100:
            insight = "Good ESG performance. Some areas can still improve."
        elif 700 <= score < 900:
            insight = "Moderate ESG performance. The company needs to work on sustainability."
        else:
            insight = "Poor ESG performance. Significant improvement is needed."
        
        return (f"Company: {company_name}\n"
                f"ESG Grade: {grade}\n"
                f"ESG Score: {score}\n"
                f"ESG Level: {level}\n"
                f"Insight: {insight}")


if __name__ == "__main__":
    file_path = "new dataset for python.csv"  
    esg_analyzer = ESGAnalyzer(file_path)
    
    
    company_name = input("Enter a company name: ")
    print(esg_analyzer.get_esg_info(company_name))
    
```

```python
# STEP 4 - DASHBOARD

import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import pandas as pd
import plotly.express as px


file_path = "new dataset for python.csv"  # Update with correct file path
df = pd.read_csv(file_path)

required_columns = ['Exchange', 'Industry', 'Total_score', 'Name']
if not all(col in df.columns for col in required_columns):
    raise ValueError("Dataset must contain 'Exchange', 'Industry', 'Total_score', and 'Name' columns")


app = dash.Dash(__name__)


app.layout = html.Div([
    html.H1("ESG Score Dashboard", style={'textAlign': 'center', 'color': '#4CAF50'}),
    
    html.Label("Select Exchange:"),
    dcc.Dropdown(
        id='exchange-dropdown',
        options=[{'label': exchange, 'value': exchange} for exchange in sorted(df['Exchange'].unique())],
        value=df['Exchange'].unique()[0],
        clearable=False
    ),
    
    html.Label("Select Industry:"),
    dcc.Dropdown(
        id='industry-dropdown',
        options=[{'label': industry, 'value': industry} for industry in sorted(df['Industry'].unique())],
        value=df['Industry'].unique()[0],
        clearable=False
    ),
    
    dcc.Graph(id='bar-chart')
])


@app.callback(
    Output('bar-chart', 'figure'),
    [Input('exchange-dropdown', 'value'), Input('industry-dropdown', 'value')]
)
def update_chart(selected_exchange, selected_industry):
    filtered_df = df[(df['Exchange'] == selected_exchange) & (df['Industry'] == selected_industry)]
    
    
    filtered_df = filtered_df.sort_values(by='Total_score')
    
    
    fig = px.bar(filtered_df, x='Total_score', y='Name', orientation='h',
                 title=f"Companies in {selected_industry} ({selected_exchange}) ranked by Total Score",
                 labels={'Total_score': 'Total ESG Score', 'Name': 'Company Name'},
                 color='Total_score', color_continuous_scale='Viridis')
    
    return fig


if __name__ == '__main__':
    app.run(debug=True, port=8052)
```
